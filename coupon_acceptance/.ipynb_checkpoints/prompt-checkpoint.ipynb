### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





### Data Description
Keep in mind that these values mentioned below are average values.

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
# Filter rows and create a copy of the resultant frame.
bar_df = cleaned_data[cleaned_data['coupon'] == 'Bar']
bar_df = bar_df.copy()
```

2. What proportion of bar coupons were accepted?

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
# To calculate the acceptance rate we first filter all the rows where the coupon was accepted 
# and then split it into 2 categories returning True where the bar visits was 3 or less
bar_filter = ['never','less1','1~3']
bar_df_lte3 = bar_df.copy()[bar_df.Y ==1]

bar_df_lte3['criteria'] = bar_df.apply(lambda row: (row.Bar in bar_filter), axis=1)
print("Acceptance Rate for people who went to bar 3 or less times is",bar_df_lte3['criteria'].sum()/len(bar_df_lte3))

fig = sns.countplot(data=bar_df_lte3, x='criteria',hue='criteria', dodge=False)
plt.title("Distribution of customers accepting the Bar coupon with visits 3 or less")
plt.savefig('images/image8.png')
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
# Let us create a dataframe for rows where coupon was accepted.
bar_df_gte1_25 = bar_df.copy()[bar_df.Y ==1]

# Let us convert the age column to numeric.
bar_df_gte1_25['age'] = bar_df_gte1_25['age'].str.replace('50plus','50')
bar_df_gte1_25['age'] = bar_df_gte1_25['age'].str.replace('below21','21')
bar_df_gte1_25['age'] = pd.to_numeric(bar_df_gte1_25['age'])

bar_filter = ['never','less1']

bar_df_gte1_25['criteria'] = bar_df_gte1_25.apply(lambda row: (row.Bar not in bar_filter) & (row.age>25), axis=1)

print("Acceptance Rate for people who meet the criteria is",bar_df_gte1_25['criteria'].sum()/len(bar_df_gte1_25))

fig = sns.countplot(data=bar_df_gte1_25, x='criteria',hue='criteria', dodge=False)
plt.title("Distribution of customers accepting the Bar coupon( critera visits > 1, age >25)")
plt.savefig('images/image9.png')
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
# Let us create a dataframe for rows where coupon was accepted.
bar_df_gte1_passanger = bar_df.copy()[bar_df.Y ==1]
# Create filter list to be used for creating the criteria
passenger_list = ['Friend(s)','Partner']
occupation_not_list = ['Farming Fishing & Forestry']
bar_gt1 = ['never','less1']
bar_df_gte1_passanger['criteria'] = bar_df_gte1_passanger.apply(lambda row: (row.passanger in passenger_list)  & (row.occupation not in occupation_not_list) & (row.Bar not in bar_gt1), axis=1)

print("Acceptance Rate for people who meet the criteria is",bar_df_gte1_passanger['criteria'].sum()/len(bar_df_gte1_passanger))


sns.countplot(data=bar_df_gte1_passanger, x='criteria',hue='criteria', dodge=False)

plt.title("Distribution of customers accepting the Bar coupon( critera visits > 1, no kid passengers, not in occupation Farming Fishing & Forestry )")
plt.savefig('images/image10.png')
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
#go to bars more than once a month, had passengers that were not a kid, and were not widowed
# Filter the rows that accepted the coupon
bar_df_gte1_passanger = bar_df.copy()[bar_df.Y ==1]

passenger_list = ['Friend(s)','Partner']
married_not_status = ['Widowed']
bar_gt1 = ['never','less1']
# Add a True/False flag based on the required criteria 
bar_df_gte1_passanger['criteria'] = bar_df_gte1_passanger.apply(lambda row: (row.passanger in passenger_list)  & (row.maritalStatus not in married_not_status) & (row.Bar not in bar_gt1), axis=1)

print("Acceptance Rate for people who meet the criteria is",bar_df_gte1_passanger['criteria'].sum()/len(bar_df_gte1_passanger))

sns.countplot(data=bar_df_gte1_passanger, x='criteria',hue='criteria', dodge=False)
plt.title("Distribution of customers accepting the Bar coupon( critera visits > 1, no kid passengers, were not widowed )")
plt.savefig('images/image11.png')
```

```python
#go to bars more than once a month and are under the age of 30
# Filter the rows that accepted the coupon
bar_df_gte1_passanger = bar_df.copy()[bar_df.Y ==1]
# Make the age a numeric column
bar_df_gte1_passanger['age'] = bar_df_gte1_passanger['age'].str.replace('50plus','50')
bar_df_gte1_passanger['age'] = bar_df_gte1_passanger['age'].str.replace('below21','21')
bar_df_gte1_passanger['age'] = pd.to_numeric(bar_df_gte1_passanger['age'])
bar_gt1 = ['never','less1']
bar_df_gte1_passanger['criteria'] = bar_df_gte1_passanger.apply(lambda row: (row.age <30) & (row.Bar not in bar_gt1), axis=1)

print("Acceptance Rate for people who meet the criteria is",bar_df_gte1_passanger['criteria'].sum()/len(bar_df_gte1_passanger))

fig = sns.countplot(data=bar_df_gte1_passanger, x='criteria',hue='criteria', dodge=False)
plt.title("Distribution of customers accepting the Bar coupon( critera visits > 1, age <30 )")
plt.savefig('images/image12.png')
```

```python
# go to cheap restaurants more than 4 times a month and income is less than 50K.
bar_df_gte1_passanger = bar_df.copy()[bar_df.Y ==1]

cheap_restaurant =['4~8','4~8']
income_group = ['$25000 - $37499','$12500 - $24999','$37500 - $49999','Less than $12500']
bar_df_gte1_passanger['criteria'] = bar_df_gte1_passanger.apply(lambda row: (row.RestaurantLessThan20 in cheap_restaurant) & (row.income in income_group), axis=1)
print("Acceptance Rate for people who meet the criteria is",bar_df_gte1_passanger['criteria'].sum()/len(bar_df_gte1_passanger))

fig = sns.countplot(data=bar_df_gte1_passanger, x='criteria',hue='criteria', dodge=False)
plt.title("Distribution of customers accepting the Bar coupon( critera cheap restaurants > 4, income <50k )")
plt.savefig('images/image13.png')
```

Before furhter analysis converting some of the categorical columns to numerical to make it easier for analysis. 

```python
#Convert CoffeeHouse to a numerical feature, we can solve this even without conversion
coffee_house_df['CoffeeHouse'].replace(['less1', '1~3','4~8','gt8','never'],
                        [0, 3,8,9,-1], inplace=True)
coffee_house_df['CoffeeHouse'] = pd.to_numeric(coffee_house_df['CoffeeHouse'])

# Converting the time column to numeric by replacing the values with the 24 hr format.
coffee_house_df['time'].replace(['7AM', '10AM','2PM','6PM','10PM'],
                        [7, 10,14,18,20], inplace=True)
coffee_house_df['time'] = pd.to_numeric(coffee_house_df['time'])

# Converting the range of income to the upper bound for each bucket
coffee_house_df['income'].replace(['Less than $12500', '$12500 - $24999','$25000 - $37499','$37500 - $49999','$50000 - $62499','$62500 - $74999','$75000 - $87499','$87500 - $99999','$100000 or More'],
                        [12500, 24999,37499,49999,62499,74999,87499,99999,100000], inplace=True)
coffee_house_df['income'] = pd.to_numeric(coffee_house_df['income'])

# Converting the expiration time to number of hours.
coffee_house_df['expiration'].replace(['2h', '1d'],
                        [2, 24], inplace=True)
coffee_house_df['expiration'] = pd.to_numeric(coffee_house_df['expiration'])
```

Let us compare the acceptance rate between those who went to a Coffee 3 or fewer times a month to those who went more.

```python
df = coffee_house_df.copy()[coffee_house_df.Y ==1]
df['criteria'] = df.apply(lambda row: (row.CoffeeHouse <= 3) , axis=1)
print("Acceptance Rate for people who meet the criteria is",df['criteria'].sum()/len(df))

fig = sns.countplot(data=df, x='criteria',hue='criteria', dodge=False)
plt.title("Distribution of customers accepting the coupon( critera visits <= 4)")
plt.savefig('images/image14.png')
```

From the above image customers who visit less frequently are likely to use coupons that those who visit more frequently.
